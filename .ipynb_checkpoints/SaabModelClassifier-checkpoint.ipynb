## View Data

```python
batch_tfms = [*aug_transforms(size=224, max_warp=0), Normalize.from_stats(*imagenet_stats)]
item_tfms = RandomResizedCrop(460, min_scale=0.85, ratio=(1.,1.))
bs=128
```

```python
def multi_l(x): return [x]
```

```python
# Create Datablock
cars = DataBlock(blocks=(ImageBlock, MultiCategoryBlock),
                  get_items=get_image_files,
                  splitter=RandomSplitter(0.1),                  
                  get_y=[RegexLabeller(pat = r'/(?<=\/)(.*)(?=\/)'), multi_l],                  
                  item_tfms=item_tfms,
                  batch_tfms=batch_tfms)
```

```python
dls = cars.dataloaders(DATA_PATH,  bs=bs)
```

```python
# Good! Let's take a look at some of our pictures then.
dls.show_batch(max_n=9)
```

```python
dls.vocab
```

# Train the model

## First Attempt with xresnet34

```python
learn = cnn_learner(dls, resnet34, pretrained=True, metrics=[partial(accuracy_multi, thresh=0.8)],
                   loss_func=BCEWithLogitsLossFlat())
```

```python
learn.fit_one_cycle(3)
```

```python
learn.save('stage-1')
```

Test learner with a Non-Saab Image. Let's try a bike. The model should output nothing because it is not trained for classifying bike-images

```python
im = PILImage.create('bike.jpg')
im.show()
```

```python
learn.predict('bike.jpg')
```

```python
learn.load('stage-1')
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(15, lr_max=1e-2)
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(20, lr_max=slice(1e-5, 1e-2/5))
```

```python
learn.save('stage-2')
```

## Interpretation

```python
learn.load('stage-2')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(3, figsize=(9,9))
```

```python
interp.most_confused(min_val=20)
```

## Train again with ResNet50

```python
bs = 32
dls = cars.dataloaders(DATA_PATH,  bs=bs)
```

```python
learn = cnn_learner(dls, resnet101, pretrained=True, metrics=[partial(accuracy_multi, thresh=0.8)],
                   loss_func=BCEWithLogitsLossFlat())
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('resnet50')
```

```python
learn.load('resnet50')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(3, figsize=(9,9))
```
